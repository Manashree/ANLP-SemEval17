```python
import pandas as pd
import numpy as np
import nltk
import re
import string
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from nltk.stem.wordnet import WordNetLemmatizer
import warnings

from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import SGDClassifier

from scipy.stats.stats import pearsonr

import numpy as np
import pandas as pd
from sklearn.cross_validation import KFold
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

ESTIMATORS = {
    "linear": LinearRegression(),
    "svr":  SVR(C=1.0, epsilon=0.2),
    "rfr": RandomForestRegressor(n_estimators=250),
}
```

```python
temp_values = ['MSRpar', 'MSRvid', 'SMTeuroparl', 'surprise.OnWN',
       'surprise.SMTnews', 'headlines', 'OnWN', 'FNWN', 'deft-forum',
       'deft-news', 'tweet-news', 'answers-forums', 'answers-students',
       'belief', 'images', 'answer-answer', 'plagiarism',
       'question-question', 'postediting']
```

```python
for val in temp_values:
    train_modified = pd.read_csv('final-modified.csv')
    train_data1 = train_modified[train_modified['type']==val]
    train_data1_labels = train_data1['original_score'].values
    train_data1 = train_data1.drop(['Unnamed: 0', 'year', 'type', 'sentence1', 'sentence2', 'cleaned_S1','cleaned_S2', 'cleaned_tagged_S1', 'cleaned_tagged_S2','original_score'],axis=1)
    train_data1 = train_data1.fillna(0)

    n_folds = 10
    train_predictions = pd.DataFrame({'id': range(train_data1.shape[0])})
    train_predictions.set_index(['id'])
    for name, estimator in ESTIMATORS.items():
        #print("Model: ", name)

        kf = KFold(train_data1.shape[0], n_folds=n_folds)
        hold_prediction = []
        eval_score = 0

        for i, (train_index, test_index) in enumerate(kf):
            X_train, X_val = train_data1.iloc[train_index], train_data1.iloc[test_index]
            y_train, y_val = train_data1_labels[train_index], train_data1_labels[test_index]
            estimator.fit(X_train, y_train)
            hold_prediction.extend(estimator.predict(X_val))
            #print("Finished fold",i)
        train_predictions[name] = hold_prediction
    
    print("Statistics for algorithm:",val)
    for col in ESTIMATORS.keys():
        print(col,":",pearsonr(train_predictions[col], train_data1_labels)[0])
```
